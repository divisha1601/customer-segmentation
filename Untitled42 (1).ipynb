```python
import pandas as pd

df = pd.read_csv("Mall_Customers.csv")
df.head()
```

## Dataset Understanding
The dataset contains customer information such as age, income, and spending score.
We will use Annual Income and Spending Score for clustering.

```python
df.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(
    x='Annual Income (k$)',
    y='Spending Score (1-100)',
    data=df
)

plt.title("Customer Distribution")
plt.show()
```

## Data Visualization
The scatter plot shows how customers are distributed.
We can visually observe that customers form natural groups.

```python
X = df[['Annual Income (k$)', 'Spending Score (1-100)']]
```

```python
from sklearn.cluster import KMeans

wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.title("Elbow Method")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.show()
```

## Elbow Method
The elbow method helps determine the optimal number of clusters.
From the graph, K = 5 appears to be the best choice.

```python
kmeans = KMeans(n_clusters=5, random_state=42)
y_kmeans = kmeans.fit_predict(X)
```

## Model Building
K-Means clustering is applied to group customers into 5 clusters based on similarity.

```python
plt.scatter(X.iloc[:,0], X.iloc[:,1], c=y_kmeans, cmap='viridis')

plt.scatter(
    kmeans.cluster_centers_[:,0],
    kmeans.cluster_centers_[:,1],
    s=200, c='red'
)

plt.xlabel("Income")
plt.ylabel("Spending Score")
plt.title("Customer Segments")
plt.show()
```

## Customer Segments (Insights)

Cluster 1: High income, high spending → Premium customers  
Cluster 2: High income, low spending → Potential customers  
Cluster 3: Low income, high spending → Impulsive buyers  
Cluster 4: Low income, low spending → Budget customers  
Cluster 5: Medium income, medium spending → Average customers
